```python
import matplotlib.pyplot as plt
import numpy as np
\
#Data
x = np.arange(0,10,0.1)
# [0,0.1,0.2,0.3,..........9.9]
y = np.sin(x)


#Graph creation
plt.plot(x,y)

#Detemine chart properties
plt.title("sinüs Grafiği")
plt.xlabel("x eksen")
plt.xlabel("y eksen")

plt.show()
```

```python
from numpy.ma.core import size
from matplotlib import colors
import matplotlib.pyplot as plt
import numpy as np

#Data 
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
size = np.random.randint(50,150,50)

plt.scatter(x, y, c=colors , s=size)

plt.title("Nokta grafigi")
plt.xlabel("x eksen")
plt.xlabel("y lapel")

plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

#Data 
data = np.random.randn(1000)

plt.hist(data, bins=30)

plt.title("Histogram")
plt.xlabel("Degerler")
plt.ylabel("Frequency")

plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

#Data
x = ['A','B','C','D','E']
y = np.random.randint(1,10,5)

plt.bar(x,y)

plt.title("sinüs Grafiği")
plt.xlabel("Kategoriler")
plt.ylabel("Degerler")

plt.show()
```
